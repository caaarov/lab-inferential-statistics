```python
import math
```

```python
sample_mean = 130.1
pop_mean = 120
sample_std = 21.21
n = 100
statistic = (sample_mean - pop_mean)/(sample_std/math.sqrt(n))
print("Statistic is: ", statistic)
```

```python
from scipy import stats
from numpy.random import normal


samples = {}

for i in range(10):
    sample_name = "sample_" + str(i)
    #creating a sample matching the variables above
    samples[sample_name] = normal(loc = 130.1, scale = 21.21, size = 100)
    sample_mean = "sample_" + str(i) + "_mean"
    samples[sample_mean] = np.mean(samples[sample_name])
    sample_std = "sample_" + str(i) + "_std"
    samples[sample_std] = np.std(samples[sample_name],ddof=1)
    sample_statistic = "sample_" + str(i) + "_z-statistic"
    samples[sample_statistic] = (samples[sample_mean]- pop_mean)/(samples[sample_std]/math.sqrt(n)) 
    print("The z-statistic for the sample {} is: {}".format(i,samples[sample_statistic]))
```

```python
print("Assuming a significance level of 0.05")
print()

for i in range(10):
    sample_name = "sample_" + str(i)
    # In the next line, 85 is the population's mean.
    print("The p-value of sample {} is: {:-5.3}".format(i,stats.ttest_1samp(samples[sample_name],120)[1]))
    print("The values in the sample are: ")
    print(samples[sample_name])
    sample_mean = "sample_" + str(i) + "_mean"
    print(samples[sample_mean])
    print()
    if (stats.ttest_1samp(samples[sample_name],120)[1] < 0.05 ):
        print("Therefore we discard the null hypothesis Ho, as it's very unlikely to get sample {} given Ho.".format(i))
    else: 
        print("We accept the null hypothesis Ho, as it's very likely to obtain sample {} given Ho".format(i) )
    print()
```

```python
machine.drop(["Unnamed: 0"],axis=1, inplace=True)
students.drop(["Unnamed: 0"],axis=1, inplace=True)
```

```python
machine.columns = ["new_machine", "old_machine"]
```

```python
students
```

```python
students.columns =["sophomores","juniors"]
```

```python
avg_new_machine=np.mean(machine["new_machine"])
avg_old_machine=np.mean(machine["old_machine"])
```

In a packing plant, a machine packs cartons with jars. It is supposed that a new machine will pack faster on the average than the machine currently used. To test that hypothesis, the times it takes each machine to pack ten cartons are recorded. 

```python
#machine
#HO

sample_mean1 = avg_new_machine
sample_std1 = np.std(machine["new_machine"])
n1 = 10
sample_mean2 = avg_old_machine
sample_std2 = np.std(machine["old_machine"])
n2 = 10

pooled_sample_std = math.sqrt(((n1-1)*sample_std1**2 + (n2-1)*sample_std2**2)/(n1+n2-2))
statistic = (sample_mean1-sample_mean2)/(pooled_sample_std*math.sqrt((1/n1)+(1/n2)))
print("T Statistic is: ", statistic)
```

```python
from scipy.stats import t

print("P value is: ",t.cdf(statistic,n1+n2-2))
print("Critical Value of z is: ", t.ppf(0.05, n1+n2-2)) #alpha is 0.05
```

```python
#we reject the 0HP (there is no differences), which is supporting the HP new machine is faster
```

```python
#["sophomores","juniors"]

avg_jun=np.mean(students["juniors"])
avg_sop=np.mean(students["sophomores"])
```

```python
#student

sample_mean1 = avg_jun
sample_std1 = np.std(students["juniors"])
n1 = 13
sample_mean2 = avg_sop
sample_std2 = np.std(students["sophomores"])
n2 = 17

pooled_sample_std = math.sqrt(((n1-1)*sample_std1**2 + (n2-1)*sample_std2**2)/(n1+n2-2))
statistic = (sample_mean1-sample_mean2)/(pooled_sample_std*math.sqrt((1/n1)+(1/n2)))
print("T Statistic is: ", statistic)
```

```python
from scipy.stats import t

print("P value is: ", 1-t.cdf(statistic,n1+n2-2))
print("Critical Value of z is: ", t.ppf(0.05, n1+n2-2)) #alpha is 0.05
```

```python
#HP0 (there is no differences) is not rejected
```
